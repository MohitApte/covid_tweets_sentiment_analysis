Removing Nulls And Duplicates

```python
df.drop_duplicates(subset='OriginalTweet',inplace=True)
```

```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
sw = stopwords.words('English')
print(sw)
lm = WordNetLemmatizer()
```

```python
df.reset_index(drop=True, inplace=True)
```

Preprocess the Covid tweets based on the following parameter:
a) Tokenizing words
b) Convert words to lower case
c) Removing Punctuations
d) Removing Stop words
e) Stemming or lemmatizing the words

```python
res = []
for i in df['OriginalTweet']:
    t = re.sub('(@[A-Za-z0-9]+)|(#[A-Za-z0-9]+)|([^0-9A-Za-z t])|(w+://S+)',' ',i)
    t = t.lower()
    t = word_tokenize(t)
    t = [i for i in t if i not in sw]
    t = [lm.lemmatize(i) for i in t]
    t = " ".join(t)
    res.append(t)
print(res)
```

```python
df1
```

Convert the 'Extremely Positive' and 'Extremely Negative' Sentiments to 'Positive' and
'Negative' sentiments respectively

```python
for index, row in df.iterrows():
    if df.loc[index, 'Sentiment'] == 'Extremely Positive':
        df['Sentiment'][index] = 'Positive'
    if df.loc[index, 'Sentiment'] == 'Extremely Negative':
        df['Sentiment'][index] = 'Negative'

    
```

```python
df
```

Transform the words into vectors using TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer(max_features=1500)
r3 = tf.fit_transform(res).todense()
print(len(tf.get_feature_names_out()))
```

```python
x = r3
y = df['Sentiment']
print(type(x))
print(type(y))
```

Split data into training and test data.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Multinomial Naïve Bayes Classification

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB
```

```python
m1 = MultinomialNB()
m1.fit(x_train, y_train)
```

```python
print('Training Score',m1.score(x_train, y_train))
print('Testing Score',m1.score(x_test, y_test))
```

```python
ypred_m1 = m1.predict(x_test)
print(ypred_m1)
```

Compute Confusion matrix and classification report for each of these models


```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm_m1 = confusion_matrix(y_test, ypred_m1)
print(cm_m1)
print(classification_report(y_test, ypred_m1))
```
